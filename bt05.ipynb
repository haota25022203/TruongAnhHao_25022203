```python
#b3
def so_nguyen(a):
    if a<2:
        return False
    for i in range(2,a):
        if a%i==0:
            return False
    return True
so_nguyen(9)
```

```python
#b4
def so_hoan_hao(n):
    s = 0
    for i in range(1,n):
        if n%i==0:
            s += i
    if s == n:
        return True
    else:
        return False
so_hoan_hao(6)
```

```python
#b5
def vi_tri(arr, k):
    for i in range(len(arr)):
        if arr[i] == k:
            return i+1
    return -1
vi_tri([1, 2, 3, 4, 5], 5)
```

```python
#b6
def giai_thua(n):
    s = 1
    for i in range(1,n+1):
        s *= i
    print(s)
giai_thua(3)
```

```python
#b9
def tong(n):
    n = abs(n)
    s = 0
    while n>0:
        s += n%10
        n //= 10
    print(s)
tong(123456) 
```

```python
#b11
def  is_product_divisible_by_10(a, b):
    if (a*b)%10==0:
        return True
    else:
        return False
is_product_divisible_by_10(1, 10)
```

```python
#b12
def divide_with_remainder(dividend, divisor):
    q = dividend//divisor
    r = dividend -  divisor*q
    if 0 <= abs(r) < abs(divisor):
        return(q,r)

divide_with_remainder(100, 12)
```

```python
#b13
def max_of_two(a, b):
    if a > b:
        return a
    else:
        return b
max_of_two(10, 100)
```

```python
#b14
def  compute(a, b, op):
    if op == 1:
        return a+b
    elif op == 2:
        return a-b
    elif op == 3:
        return a*b
    else:
        return a/b
compute(1, 2, 4)
```

```python
#b15
def is_right_triangle(a, b, c):
    if a>0 and b>0 and c>0:
        if a*a + b*b == c*c or  a*a == b*b + c*c or b*b == a*a + c*c:
            return True
        else:
            return False
    else:
        print("nhap lai so khac")
is_right_triangle(3, 4, 5)
```
