```python
#b1
nums = list(map(int, input().split()))

result = []

for x in nums:
    if x not in result:
        result.append(x)

print(result)
```

```python
#b2
nums = list(map(int, input().split()))

result = []
tong = 0

for x in nums:
    tong += x
    result.append(tong)

print(result)
```

```python
#b3
t = tuple(map(int, input().split()))
k = int(input())

n = len(t)
k = k % n  # tránh k > n

result = t[k:] + t[:k]

print(result)
```

```python
#b4
pairs = input().split()

result = {}

for p in pairs:
    key, value = p.split(":")
    
    value = int(value)
    
    if key not in result:
        result[key] = [value]
    else:
        result[key].append(value)

print(result)
```

```python
#b5
lst = list(map(int, input().split()))

result = {
    'positives': 0,
    'negatives': 0,
    'zeros': 0
}

for x in lst:
    if x > 0:
        result['positives'] += 1
    elif x < 0:
        result['negatives'] += 1
    else:
        result['zeros'] += 1

print(result)
```

```python
#b7
t = tuple(map(int, input().split()))

first = t[0]
last = t[-1]
reversed_t = t[::-1]

print("First:", first)
print("Last:", last)
print("Reversed:", reversed_t)
```

```python
#b8
lst = input().split()

result = {}

for s in lst:
    if s in result:
        result[s] += 1
    else:
        result[s] = 1

print(result)
```

```python
#b9
d1 = input().split()
d2 = input().split()

dict1 = {}
dict2 = {}

for item in d1:
    k, v = item.split(":")
    dict1[k] = int(v)

for item in d2:
    k, v = item.split(":")
    dict2[k] = int(v)

result = dict1.copy()

for k, v in dict2.items():
    if k in result:
        result[k] += v
    else:
        result[k] = v

for k in sorted(result):
    print(f"{k}:{result[k]}", end=" ")
```
