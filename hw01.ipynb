```python
#b3
c = input()
if 'A' <= c <= 'Z':
    ket_qua = chr(ord(c) + 32)
    print(f"Kí tự chữ thường tương ứng là: {ket_qua}")
elif 'a' <= c <= 'z':
    ket_qua = chr(ord(c) - 32)
    print(f"Kí tự chữ hoa tương ứng là: {ket_qua}")
else:
    print("Kí tự bạn nhập không phải là chữ cái.")
```

```python
#b6
a, b, c = map(int, input().split())
if a+b>c>0 and a+c>b>0 and b+c>a>0:
    p = (a+b+c)/2
    s = (p*(p-a)*(p-b)*(p-c))**0.5
    print(f'{s:.1f}')
else:
    print("khong phai 3 canh tam giac")
```
