```python
#b3
a, b, c = map(float, input().split())
if a+b>c>0 and a+c>b>0 and b+c>a>0:
    p = (a+b+c)/2
    s = (p * (p-a) * (p-b) * (p-c) )**(1/2)
    if a==b==c:
        print("tam giac deu", s, p*2)
    elif a==b or a==c or b==c:
        print("cân", s, p*2)
    elif a*a==b*b +c*c or b*b==a*a + c*c or c*c == b*b + a*a:
        print("vuông", s, p*2)
    else:
        print("thường", s, p*2)
else:
    print("khong phai tam giac") 
```
