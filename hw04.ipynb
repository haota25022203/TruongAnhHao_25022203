```python
#b18
a, b = map(int, input().split())
while a!=b:
    if a>b:
        a = a - b
    else:
        b = b-a
for i in range(1,a+1):
    if a%i==0:
        print(i, end = " ")
```

```python
#b21
n = int(input())
s = 0
while n!=0:
    s += n%10
    n //= 10
print(s)
```

```python
#b22
n = int(input())
a, b = 0, 0
while n!=0:
    i = n%10
    if i%2==0:
        a += 1
    else:
        b += 1
    n //= 10
print(a, b)
```

```python
#b23
n = int(input())
k = int(-1/2 + ((1+8*n)**(1/2))/2)
if k*(k+1)/2>=n:
    k = k - 1
print(k)
```

```python
#b24
a = int(input())
s = 0
n = 0
while s <= a:
    n += 1
    s += 1/n
print(n, s)
```

```python
#b25
max = int(input())
min = max
while True:
    n = int(input())
    if n == -1:
        break
    if n>max:
        max = n
    if n<min:
        min = n
print(max, min)
```

```python
#b29
s1 = input()
s2 = ''
for i in s1:
    s = ord(i)
    if s>96:
        s2 = s2 + chr(s-32)
    else:
        s2 = s2 + chr(s+32)
print(s2)
```

```python
#b31
s1 = str(input())
h, t, s = 0, 0, 0
for i in s1:
    n = ord(i)
    if n>48 and n<58:
        s += 1
    elif n>64 and n<91:
        h += 1
    else:
        t += 1
print(h, t, s)
```
