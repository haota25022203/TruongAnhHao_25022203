```python
#b16
def is_armstrong(x):
    # chuyển số thành chuỗi để đếm chữ số
    s = str(x)
    n = len(s)
    
    tong = 0
    for digit in s:
        tong += int(digit) ** n
    
    return tong == x
x = int(input("Nhập số: "))

# kiểm tra
if is_armstrong(x):
    print("Là số Armstrong")
else:
    print("Không phải số Armstrong")
```

```python
#b17
def is_perfect_square(x):
    if x < 0:
        return False
    
    i = 0
    while i * i <= x:
        if i * i == x:
            return True
        i += 1
    
    return False

# nhập số
x = int(input("Nhập số: "))

# kiểm tra
if is_perfect_square(x):
    print("Là số chính phương")
else:
    print("Không phải số chính phương")
```

```python
#b18
def is_palindrome(x):
    if x < 0:
        return False
    
    temp = x
    dao = 0
    
    while temp > 0:
        digit =7
        temp % 10
        dao = dao * 10 + digit
        temp //= 10
    
    return dao == x

# nhập số
x = int(input("Nhập số: "))

# kiểm tra
if is_palindrome(x):
    print("Là số đối xứng")
else:
    print("Không phải số đối xứng")
```

```python
#b19
def is_magic_number(x):
    if x <= 0:
        return False
    
    while x > 9:
        tong = 0
        while x > 0:
            tong += x % 10
            x //= 10
        x = tong
    
    return x == 1

# nhập số
x = int(input("Nhập số: "))

# kiểm tra
if is_magic_number(x):
    print("Là số kỳ diệu")
else:
    print("Không phải số kỳ diệu")
```

```python
#b23
def quadratic_roots_count(a, b, c):
    if a == 0:
        if b == 0:
            return 0
        else:
            return 1
    
    delta = b * b - 4 * a * c
    
    if delta > 0:
        return 2
    elif delta == 0:
        return 1
    else:
        return 0

# nhập hệ số
a = float(input("Nhập a: "))
b = float(input("Nhập b: "))
c = float(input("Nhập c: "))

# in kết quả
print(quadratic_roots_count(a, b, c))
```

```python
#b24
def are_collinear(x1, y1, x2, y2, x3, y3):
    return (x2 - x1) * (y3 - y1) == (x3 - x1) * (y2 - y1)


x1, y1, x2, y2, x3, y3 = map(float, input("Nhập 6 số: ").split())

# kiểm tra
if are_collinear(x1, y1, x2, y2, x3, y3):
    print("Ba điểm thẳng hàng")
else:
    print("Ba điểm không thẳng hàng")
```

```python
#b25
def area(x1, y1, x2, y2, x3, y3):
    return abs(x1*(y2 - y3) + x2*(y3 - y1) + x3*(y1 - y2))

def point_in_triangle(A, B, C, P):
    x1, y1 = A
    x2, y2 = B
    x3, y3 = C
    x, y = P
    
    S = area(x1, y1, x2, y2, x3, y3)
    S1 = area(x, y, x2, y2, x3, y3)
    S2 = area(x1, y1, x, y, x3, y3)
    S3 = area(x1, y1, x2, y2, x, y)
    
    return S == S1 + S2 + S3

def point_in_quadrilateral(quad, p):
    A, B, C, D = quad
    return point_in_triangle(A, B, C, p) or point_in_triangle(A, C, D, p)

# nhập dữ liệu (1 dòng)
# ví dụ: 0 0 4 0 4 3 0 3 2 1
data = list(map(float, input("Nhập 10 số: ").split()))

quad = [(data[0], data[1]), (data[2], data[3]), 
        (data[4], data[5]), (data[6], data[7])]
p = (data[8], data[9])

# kiểm tra
if point_in_quadrilateral(quad, p):
    print("True")
else:
    print("False")
```

```python
#b27
def count_nonzero_digits(lst):
    count = 0
    
    for num in lst:
        num = abs(num)
        if num == 0:
            continue
        
        while num > 0:
            digit = num % 10
            if digit != 0:
                count += 1
            num //= 10
    
    return count

# nhập list trên 1 dòng
lst = list(map(int, input("Nhập các số: ").split()))

# in kết quả
print(count_nonzero_digits(lst))
```

```python
#b28
import math

def angle_ABC(x1, y1, x2, y2, x3, y3):
    BAx = x1 - x2
    BAy = y1 - y2
    BCx = x3 - x2
    BCy = y3 - y2
    
    dot = BAx * BCx + BAy * BCy
    
    len_BA = math.sqrt(BAx**2 + BAy**2)
    len_BC = math.sqrt(BCx**2 + BCy**2)
    
    if len_BA == 0 or len_BC == 0:
        return 0
    
    cos_theta = dot / (len_BA * len_BC)
    cos_theta = max(-1, min(1, cos_theta))
    
    return math.degrees(math.acos(cos_theta))

# nhập 1 dòng
x1, y1, x2, y2, x3, y3 = map(float, input("Nhập 6 số: ").split())

print(angle_ABC(x1, y1, x2, y2, x3, y3))
```

```python
#b29
def max_min(a, b, c):
    max_val = a
    min_val = a
    
    if b > max_val:
        max_val = b
    if c > max_val:
        max_val = c
    
    if b < min_val:
        min_val = b
    if c < min_val:
        min_val = c
    
    return [max_val, min_val]

a, b, c = map(float, input("Nhập a b c: ").split())

print(max_min(a, b, c))
```

```python
#b30
def combination(n, k):
    if k < 0 or k > n:
        return 0
    
    res = 1
    for i in range(1, k + 1):
        res = res * (n - i + 1) // i
    
    return res
n, k = map(int, input("Nhập n k: ").split())

print(combination(n, k))
```

```python
#b34
def has_distinct_digits(x):
    seen = []
    
    while x > 0:
        digit = x % 10
        if digit in seen:
            return False
        seen.append(digit)
        x //= 10
    
    return True

def sum_1_to_n(n):
    if n <= 0:
        return 0
    
    total = 0
    for x in range(1, n + 1):
        if has_distinct_digits(x):
            total += x
    
    return total

# nhập
n = int(input("Nhập n: "))

print(sum_1_to_n(n))
```
